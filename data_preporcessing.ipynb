```python
import os
import pandas as pd
from os import listdir
from os.path import isfile, join

os.chdir("/content/drive/MyDrive/NHS Hospital Admissions")
```

```python
pwd = os.getcwd()
```

```python
folders = {
"Primary_diagnosis_summary": 1,
"Primary_diagnosis_3-char": 2,
"Primary_diagnosis_4-char": 3,
"All_diagnoses_3-char": 4,
"All_diagnoses_4-char": 5
}
```

```python
excel_files = [f for f in listdir(pwd) if f.endswith('.xlsx') & isfile(join(pwd, f))]
```

```python
excel_files
```

```python
tempdf = pd.read_excel(io=excel_files[0],sheet_name=1, header=10, skipfooter=13)
tempdf.insert(0, 'filename', excel_files[0])
```

```python
for folder, sheet in folders.items():
  for f in excel_files:
    print(f)
    print(sheet)
    tempdf = pd.read_excel(io=f,sheet_name=sheet, header=0, skipfooter=0, keep_default_na=False)

    # tempdf = tempdf.dropna(axis=1, how='all')
    # print(tempdf.iloc[:, 0].isnull().all())

    header = tempdf[tempdf.iloc[:, 0].str.contains(".*([Pp]rimary|[Aa]ll) [Dd]iagnos.*", regex=True)].index
    # print(header)
    if len(header) > 0:
      tempdf.drop(tempdf.index[:header[-1]], inplace=True)
      tempdf.dropna(inplace=True)
      tempdf.columns = tempdf.iloc[0]
      tempdf = tempdf[1:].reset_index(drop=True)

    footer = tempdf[tempdf.iloc[:, 0].str.contains(".*Z9.*", regex=True)].index
    # print(footer)
    if len(footer) > 0:
      tempdf.drop(tempdf.index[footer[-1]+1:], inplace=True)

    if len(header) == 0:
    #   print(tempdf.iloc[:, 0])
    # else:
      print('FIX:')
      print(join(pwd, folder, f.replace('xlsx', 'csv')))

    tempdf.insert(0, 'filename', f)
    tempdf.to_csv(join(pwd, folder, f.replace('xlsx', 'csv')))
    # print(join(folder, f.replace('xlsx', 'csv')))
```
